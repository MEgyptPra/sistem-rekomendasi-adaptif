```python
# ===== CELL 1: PRODUCTION-GRADE SETUP =====
!pip install sqlalchemy psycopg2-binary nest_asyncio asyncpg tenacity scikit-learn matplotlib seaborn pandas numpy tabulate

import nest_asyncio
nest_asyncio.apply()

import sys
import os

# ===== KONEKSI KE KODE PRODUKSI =====
# Sesuaikan path ini dengan struktur project Anda
PRODUCTION_CODE_PATH = '../pariwisata-recommender/backend'
sys.path.append(PRODUCTION_CODE_PATH)

# Import model dan service dari kode produksi
try:
    from app.models.rating import Rating
    from app.models.destination import Destination
    from app.models.category import Category
    from app.models.destination_category import DestinationCategory
    from app.services.collaborative_recommender import CollaborativeRecommender
    from app.services.content_based_recommender import ContentBasedRecommender
    from app.services.hybrid_recommender import HybridRecommender
    from app.services.mab_optimizer import MABOptimizer
    
    print("âœ… Berhasil mengimpor kode produksi!")
    PRODUCTION_MODE = True
    
except ImportError as e:
    print(f"âš ï¸ Tidak bisa mengimpor kode produksi: {e}")
    print("âŒ EVALUASI TIDAK VALID: Hubungkan dengan kode produksi terlebih dahulu!")
    print("ðŸ“ Pastikan path ini benar: '../pariwisata-recommender/backend'")
    PRODUCTION_MODE = False
    
    # Fallback minimal untuk testing
    class Rating: pass
    class Destination: pass  
    class Category: pass
    class DestinationCategory: pass

# Core imports
import asyncio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import logging
import random
from collections import Counter
import scipy.stats as stats
import math
from tabulate import tabulate
import traceback
import gc
import copy

# ML & Database
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import NMF
from sqlalchemy.ext.asyncio import create_async_engine, AsyncSession
from sqlalchemy.orm import sessionmaker
from sqlalchemy import select, text
import contextlib
from tenacity import retry, stop_after_attempt, wait_exponential

# Konfigurasi
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logger = logging.getLogger('evaluasi_notebook')

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# Database setup - SAMA DENGAN PRODUCTION
DATABASE_URL = "postgresql+asyncpg://user:rekompari@localhost:5432/pariwisata"
engine = create_async_engine(DATABASE_URL, echo=False, future=True, pool_size=5, max_overflow=5)
AsyncSessionLocal = sessionmaker(engine, expire_on_commit=False, class_=AsyncSession)

@contextlib.asynccontextmanager
async def get_db():
    db = AsyncSessionLocal()
    try:
        yield db
    finally:
        await db.close()

db_semaphore = asyncio.Semaphore(5)

@retry(stop=stop_after_attempt(3), wait=wait_exponential(multiplier=1, min=1, max=10))
async def safe_db_operation(func, *args, **kwargs):
    async with db_semaphore:
        try:
            return await func(*args, **kwargs)
        except Exception as e:
            logger.error(f"Database error: {e}")
            raise

print(f"ðŸŽ¯ Mode: {'PRODUCTION' if PRODUCTION_MODE else 'DEVELOPMENT'}")

# ===== CELL 2: PRODUCTION DATA LOADING =====
async def load_production_ratings():
    """Load data ratings dari database produksi"""
    try:
        async with get_db() as db:
            result = await db.execute(select(Rating))
            ratings = result.scalars().all()
            
            data = []
            for rating in ratings:
                data.append({
                    'user_id': rating.user_id,
                    'destination_id': rating.destination_id, 
                    'rating': float(rating.rating),
                    'created_at': rating.created_at
                })
            
            df = pd.DataFrame(data)
            print(f"âœ… Loaded {len(df)} production ratings")
            return df
            
    except Exception as e:
        print(f"âŒ Gagal load production ratings: {e}")
        raise

async def load_production_destinations():
    """Load data destinasi dari database produksi"""
    try:
        async with get_db() as db:
            # Load destinations dengan categories
            query = text("""
                SELECT 
                    d.id as destination_id,
                    d.name as destination_name,
                    d.description,
                    d.location,
                    COALESCE(c.name, 'Umum') as category_name,
                    c.id as category_id
                FROM destinations d
                LEFT JOIN destination_categories dc ON d.id = dc.destination_id
                LEFT JOIN categories c ON c.id = dc.category_id
                ORDER BY d.id
            """)
            
            result = await db.execute(query)
            rows = result.fetchall()
            
            item_features = {}
            for row in rows:
                destination_id = row[0]
                name = row[1]
                description = row[2]
                location = row[3] 
                category = row[4]
                category_id = row[5]
                
                if destination_id not in item_features:
                    item_features[destination_id] = {
                        'name': name,
                        'description': description or '',
                        'location': location or '',
                        'category': category,
                        'category_id': category_id,
                        'categories': set()
                    }
                
                item_features[destination_id]['categories'].add(category)
            
            # Convert sets to lists
            for dest_id in item_features:
                item_features[dest_id]['categories'] = list(item_features[dest_id]['categories'])
            
            print(f"âœ… Loaded {len(item_features)} production destinations with categories")
            return item_features
            
    except Exception as e:
        print(f"âŒ Gagal load production destinations: {e}")
        raise

# Load production data
try:
    ratings_df = await safe_db_operation(load_production_ratings)
    item_features = await safe_db_operation(load_production_destinations)
    
    print("ðŸ“Š Production Data Summary:")
    print(f"   Ratings: {len(ratings_df)}")
    print(f"   Users: {ratings_df['user_id'].nunique()}")
    print(f"   Destinations: {ratings_df['destination_id'].nunique()}")
    print(f"   Destination features: {len(item_features)}")
    
except Exception as e:
    print(f"âŒ CRITICAL: Tidak bisa load production data: {e}")
    print("Evaluasi tidak dapat dilanjutkan tanpa data produksi!")
    raise

# ===== CELL 3: PRODUCTION MODEL INTEGRATION =====
class ProductionIntegratedEvaluator:
    """Evaluator yang menggunakan model dari production code"""
    
    def __init__(self, train_df, test_df, item_features):
        self.train_df = train_df
        self.test_df = test_df
        self.item_features = item_features
        self.total_items = len(item_features)
        
        # Calculate item popularity dari data training
        self.item_popularity = train_df['destination_id'].value_counts().to_dict()
        self.total_ratings = len(train_df)
        
        # Available categories
        self.available_categories = set()
        for features in item_features.values():
            if 'category' in features:
                self.available_categories.add(features['category'])
    
    def precision_at_k(self, recs_ids, ground_truth_ids, k=10):
        if not recs_ids or k == 0:
            return 0.0
        recs_ids = recs_ids[:k]
        return len(set(recs_ids) & set(ground_truth_ids)) / min(k, len(recs_ids))
    
    def recall_at_k(self, recs_ids, ground_truth_ids, k=10):
        if not recs_ids or not ground_truth_ids:
            return 0.0
        recs_ids = recs_ids[:k]
        return len(set(recs_ids) & set(ground_truth_ids)) / len(ground_truth_ids)
    
    def ndcg_at_k(self, recs_ids, ground_truth_ids, k=10):
        if not recs_ids or not ground_truth_ids:
            return 0.0
            
        recs_ids = recs_ids[:k]
        
        # Calculate DCG
        dcg = 0.0
        for i, did in enumerate(recs_ids):
            if did in ground_truth_ids:
                dcg += 1 / np.log2(i + 2)
                
        # Calculate IDCG
        ideal_hits = min(len(ground_truth_ids), k)
        idcg = sum(1 / np.log2(i + 2) for i in range(ideal_hits))
        
        return dcg / idcg if idcg > 0 else 0.0
    
    def intra_list_diversity(self, recs_ids, k=10):
        if len(recs_ids) < 2:
            return 0.0
            
        recs_ids = recs_ids[:k]
        
        # Count categories
        category_counts = {}
        for item_id in recs_ids:
            if item_id in self.item_features:
                category = self.item_features[item_id].get('category', 'Unknown')
                category_counts[category] = category_counts.get(category, 0) + 1
        
        # Calculate diversity as inverse of category concentration
        total = sum(category_counts.values())
        if total == 0:
            return 0.0
            
        category_proportions = [count / total for count in category_counts.values()]
        
        # Gini-based diversity (1 - Gini coefficient)
        if len(category_proportions) > 1:
            category_proportions.sort()
            n = len(category_proportions)
            index = np.arange(1, n + 1)
            gini = (np.sum((2 * index - n - 1) * np.array(category_proportions))) / (n * np.sum(category_proportions))
            return 1 - gini
        else:
            return 0.0
    
    def coverage_at_k(self, all_recommendations, k=10):
        all_recommended_items = set()
        
        for recs in all_recommendations:
            all_recommended_items.update(recs[:k])
            
        return len(all_recommended_items) / self.total_items
    
    def novelty(self, recs_ids, k=10):
        if not recs_ids:
            return 0.0
            
        recs_ids = recs_ids[:k]
        novelty_score = 0.0
        valid_items = 0
        
        for item_id in recs_ids:
            popularity = self.item_popularity.get(item_id, 0)
            popularity_norm = popularity / self.total_ratings if self.total_ratings > 0 else 0
            if popularity_norm > 0:
                novelty_score += -math.log2(popularity_norm)
                valid_items += 1
        
        return novelty_score / valid_items if valid_items > 0 else 0.0
    
    def calculate_diversity_reward(self, recs_ids, ground_truth_ids, k=10):
        precision = self.precision_at_k(recs_ids, ground_truth_ids, k)
        diversity = self.intra_list_diversity(recs_ids, k)
        novelty = self.novelty(recs_ids, k)

        # Composite reward dengan emphasis on diversity
        reward = (0.1 * precision + 0.7 * diversity + 0.2 * novelty)
        return max(reward, 0.01)

# ===== CELL 4: PRODUCTION MODEL INITIALIZATION =====
print("ðŸš€ Initializing production models...")

# Split data untuk evaluasi
train_df, test_df = train_test_split(
    ratings_df, 
    test_size=0.2,
    random_state=42,
    stratify=ratings_df['user_id']
)

# Initialize production models
try:
    if PRODUCTION_MODE:
        # Gunakan model dari production code
        cf_model = CollaborativeRecommender()
        cb_model = ContentBasedRecommender()
        hybrid_model = HybridRecommender()
        mab_optimizer = MABOptimizer()
        
        print("âœ… Production models loaded successfully!")
    else:
        # Fallback - ini seharusnya tidak terjadi jika production mode aktif
        raise ImportError("Production models not available")
        
except Exception as e:
    print(f"âš ï¸ Error loading production models: {e}")
    print("âŒ Cannot continue without production models!")
    raise

# Initialize evaluator
evaluator = ProductionIntegratedEvaluator(train_df, test_df, item_features)

# Prepare ground truth untuk evaluasi
test_truth = test_df.groupby('user_id')['destination_id'].apply(list).to_dict()
eligible_users = [uid for uid in test_truth.keys() if len(test_truth[uid]) >= 2]
eligible_users = eligible_users[:50]  # Batasi untuk testing

ground_truth_cache = {}
for user_id in eligible_users:
    ground_truth_cache[user_id] = test_truth.get(user_id, [])

print(f"ðŸŽ¯ Evaluation Setup:")
print(f"   Training ratings: {len(train_df)}")
print(f"   Test ratings: {len(test_df)}") 
print(f"   Eligible users: {len(eligible_users)}")
print(f"   Ground truth items: {sum(len(v) for v in ground_truth_cache.values())}")

# ===== CELL 5: PRODUCTION-STYLE EVALUATION =====
async def production_evaluation_pipeline():
    """Pipeline evaluasi yang mirror production environment"""
    print("ðŸ”¥ Starting production-style evaluation...")
    
    # Contexts yang realistic untuk production
    production_contexts = [
        {'day_type': 'weekend', 'weather': 'cerah', 'time_of_day': 'siang'},
        {'day_type': 'weekday', 'weather': 'cerah', 'time_of_day': 'sore'}, 
        {'day_type': 'weekend', 'weather': 'hujan', 'time_of_day': 'pagi'},
        {'day_type': 'weekday', 'weather': 'mendung', 'time_of_day': 'malam'}
    ]
    
    all_results = []
    
    for user_id in tqdm(eligible_users, desc="Production Evaluation"):
        try:
            # Random context seperti di production
            context = random.choice(production_contexts)
            
            # Get recommendations dari production models
            # Note: Di production, ini akan dipanggil via API/service
            cf_recs = await cf_model.predict(user_id, db=await get_db().__aenter__())
            cb_recs = await cb_model.recommend_for_user(user_id, limit=20)
            hybrid_recs = await hybrid_model.get_recommendations(
                user_id=user_id, 
                context=context,
                limit=20
            )
            
            # Apply MAB optimization seperti di production
            mab_recs = await mab_optimizer.optimize_recommendations(
                user_id=user_id,
                base_recommendations=hybrid_recs,
                context=context
            )
            
            # Extract destination IDs
            cf_ids = [r.destination_id for r in cf_recs] if hasattr(cf_recs[0], 'destination_id') else [r['destination_id'] for r in cf_recs]
            cb_ids = [r.destination_id for r in cb_recs] if hasattr(cb_recs[0], 'destination_id') else [r['destination_id'] for r in cb_recs]
            hybrid_ids = [r.destination_id for r in hybrid_recs] if hasattr(hybrid_recs[0], 'destination_id') else [r['destination_id'] for r in hybrid_recs]
            mab_ids = [r.destination_id for r in mab_recs] if hasattr(mab_recs[0], 'destination_id') else [r['destination_id'] for r in mab_recs]
            
            # Store results
            all_results.append({
                'user_id': user_id,
                'context': context,
                'recommendations_cf': cf_ids[:10],
                'recommendations_cb': cb_ids[:10], 
                'recommendations_hybrid': hybrid_ids[:10],
                'recommendations_mab_optimized': mab_ids[:10],
                'ground_truth': ground_truth_cache.get(user_id, [])
            })
            
        except Exception as e:
            print(f"âš ï¸ Error evaluating user {user_id}: {e}")
            continue
    
    evaluation_df = pd.DataFrame(all_results)
    print(f"âœ… Production evaluation completed! Processed {len(evaluation_df)} users")
    
    return evaluation_df

# Jalankan evaluasi production-style
evaluation_df = await production_evaluation_pipeline()

# ===== CELL 6: PRODUCTION METRICS CALCULATION =====
def calculate_production_metrics(evaluation_df, evaluator):
    """Hitung metrics yang relevant untuk production"""
    
    model_columns = ['recommendations_cf', 'recommendations_cb', 'recommendations_hybrid', 'recommendations_mab_optimized']
    metrics_results = {}
    
    for model_col in model_columns:
        model_name = model_col.replace('recommendations_', '')
        print(f"ðŸ“Š Calculating production metrics for {model_name}...")
        
        precision_scores = []
        recall_scores = [] 
        ndcg_scores = []
        diversity_scores = []
        novelty_scores = []
        coverage_items = []
        
        for _, row in evaluation_df.iterrows():
            recs = row[model_col]
            ground_truth = row['ground_truth']
            
            # Basic accuracy metrics
            precision_scores.append(evaluator.precision_at_k(recs, ground_truth, 10))
            recall_scores.append(evaluator.recall_at_k(recs, ground_truth, 10))
            ndcg_scores.append(evaluator.ndcg_at_k(recs, ground_truth, 10))
            
            # Business metrics
            diversity_scores.append(evaluator.intra_list_diversity(recs, 10))
            novelty_scores.append(evaluator.novelty(recs, 10))
            coverage_items.extend(recs[:10])
        
        # Aggregate metrics
        coverage = len(set(coverage_items)) / evaluator.total_items
        
        metrics_results[model_name] = {
            'precision@10': np.mean(precision_scores),
            'recall@10': np.mean(recall_scores),
            'ndcg@10': np.mean(ndcg_scores),
            'diversity': np.mean(diversity_scores),
            'novelty': np.mean(novelty_scores),
            'coverage@10': coverage,
            'num_users': len(precision_scores)
        }
        
        print(f"   âœ… {model_name}: Precision={metrics_results[model_name]['precision@10']:.4f}, "
              f"Diversity={metrics_results[model_name]['diversity']:.4f}")
    
    return metrics_results

# Calculate production metrics
production_metrics = calculate_production_metrics(evaluation_df, evaluator)

# ===== CELL 7: PRODUCTION RESULTS VISUALIZATION =====
def plot_production_results(metrics_results):
    """Visualisasi hasil production evaluation"""
    
    models = list(metrics_results.keys())
    
    # Key business metrics
    business_metrics = ['precision@10', 'recall@10', 'ndcg@10', 'diversity', 'novelty', 'coverage@10']
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.flatten()
    
    colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D']
    
    for i, metric in enumerate(business_metrics):
        values = [metrics_results[model][metric] for model in models]
        
        bars = axes[i].bar(models, values, color=colors[:len(models)], alpha=0.8, edgecolor='black')
        axes[i].set_title(f'Production {metric.upper()}', fontsize=14, fontweight='bold')
        axes[i].set_ylabel(metric)
        axes[i].tick_params(axis='x', rotation=45)
        
        # Add value labels
        for bar, value in zip(bars, values):
            height = bar.get_height()
            axes[i].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                       f'{value:.3f}', ha='center', va='bottom', fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    # Production Summary Table
    print("\n" + "="*80)
    print("PRODUCTION EVALUATION SUMMARY")
    print("="*80)
    
    summary_data = []
    for model in models:
        row = [model]
        for metric in business_metrics:
            row.append(f"{production_metrics[model][metric]:.4f}")
        row.append(production_metrics[model]['num_users'])
        summary_data.append(row)
    
    headers = ['Model'] + [m.upper() for m in business_metrics] + ['Users']
    print(tabulate(summary_data, headers=headers, tablefmt='grid'))
    
    # Business Insights
    print("\n" + "="*50)
    print("BUSINESS INSIGHTS")
    print("="*50)
    
    best_model = max(models, key=lambda x: production_metrics[x]['ndcg@10'])
    best_diversity = max(models, key=lambda x: production_metrics[x]['diversity'])
    
    print(f"ðŸŽ¯ Best Accuracy (NDCG): {best_model} ({production_metrics[best_model]['ndcg@10']:.4f})")
    print(f"ðŸŒ Best Diversity: {best_diversity} ({production_metrics[best_diversity]['diversity']:.4f})")
    print(f"ðŸ“ˆ Coverage Range: {min(production_metrics[m]['coverage@10'] for m in models):.3f} - {max(production_metrics[m]['coverage@10'] for m in models):.3f}")
    
    # Recommendation untuk production deployment
    if production_metrics['mab_optimized']['ndcg@10'] > production_metrics['hybrid']['ndcg@10']:
        print("âœ… RECOMMENDATION: Deploy MAB-optimized model to production")
    else:
        print("âœ… RECOMMENDATION: Deploy Hybrid model to production")

# Plot production results
plot_production_results(production_metrics)

# Save production evaluation results
evaluation_df.to_csv('production_evaluation_results.csv', index=False)
print("\nðŸ’¾ Production evaluation results saved to 'production_evaluation_results.csv'")

print("ðŸŽ‰ PRODUCTION EVALUATION COMPLETED SUCCESSFULLY!")
```
